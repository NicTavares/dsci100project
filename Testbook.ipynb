```r
total_data = read_csv("./matchData.csv")
```

```r
data_less_columns = select(total_data, tourney_level, winner_name, loser_name, round, best_of, minutes,  w_ace, l_ace)

novak_data = filter(data_less_columns, winner_name == "Novak Djokovic" | loser_name == "Novak Djokovic")
nadal_data = filter(data_less_columns, winner_name == "Rafael Nadal" | loser_name == "Rafael Nadal")
fed_data = filter(data_less_columns, winner_name == "Roger Federer" | loser_name == "Roger Federer")
```

```r
#mutating the novak data
novak_data = mutate(novak_data, result = ifelse(winner_name == "Novak Djokovic", 'w', 'l'))
novak_data = mutate(novak_data, aces = ifelse(result == "w", w_ace, l_ace))
novak_data = select(novak_data, -w_ace, -l_ace, -winner_name, -loser_name)

novak_data = mutate(novak_data, tourney_level = ifelse(tourney_level == "D", 1, 
                                                       ifelse(tourney_level == "F", 2, 
                                                              ifelse(tourney_level == "S", 3, 
                                                                     ifelse(tourney_level == "C", 4,
                                                                           ifelse(tourney_level == "A", 5, 
                                                                                 ifelse(tourney_level == "M", 6,
                                                                                       ifelse(tourney_level == "G", 7, NA))))))))
```

```r
novak_data = mutate(novak_data, round = ifelse(round == "R128", 1, 
                                              ifelse(round == "R64", 2, 
                                                    ifelse(round == "R32", 3, 
                                                          ifelse(round == "R16", 4, 
                                                                ifelse(round == "QF", 5, 
                                                                      ifelse(round == "SF", 6, 
                                                                            ifelse(round == "F", 7, NA))))))))
head(novak_data)
```
